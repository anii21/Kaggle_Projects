# **Function To read Text Files**

```python
import gensim
from gensim.models import Word2Vec,KeyedVectors
import os   #to deal with textual files
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess

story = []
for filename in os.listdir('/kaggle/input/game-of-thrones-books'):
    file_path = os.path.join('/kaggle/input/game-of-thrones-books',filename)
    with open(file_path,'r',encoding='ISO-8859-1') as f:
        corpus =f.read()
        raw_sent = sent_tokenize(corpus)
        for sent in raw_sent: 
            story.append(simple_preprocess(sent))
```

```python
story  #jump to model for code
```

# **Model**

```python
model = gensim.models.Word2Vec( window=10, min_count=2)
```

```python
model.build_vocab(story)
```

```python
model.train(story, total_examples=model.corpus_count, epochs=model.epochs)
```

```python
model.wv.most_similar('daenerys') #gives words which are more related with daenerys
```

```python
model.wv.most_similar('jon') 
```

```python
model.wv.doesnt_match(['jon','rikn','robb','arya','sansa','bran']) #to find different: here except jon all others are siblings by blood
```

```python
model.wv.doesnt_match(['cersei', 'jaime', 'bronn', 'tyrion'])
```

```python
model.wv['king'] #all words are representes as an 100-D array with values b/w (-3,3)
```

```python
model.wv.similarity('arya','sansa') #gives value close to 1 coz they both are sisters
```

# **Reducing the dimensionality of vectors of words to see their 3D representation**

```python
from sklearn.decomposition import PCA
```

```python
pca =PCA(n_components=3) 
```

```python
X= pca.fit_transform(model.wv.get_normed_vectors())
```

```python
X.shape
```

```python
import plotly.express as px
fig = px.scatter_3d(X[300:400],x=0,y=1,z=2,color=y[300:400])
fig.show()
```

> the above 3D representation shows us the relation between different words in the books
